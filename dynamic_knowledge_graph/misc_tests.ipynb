# Investigate list removal

```python
freq_symptom = 'Bone pain of left thumb'
freq_symptom = freq_symptom.lower()
symptom_list = ['bone pain of left thumb', 
                'bone pain of proximal phalanx of left thumb', 
                'bone pain of distal phalanx of left thumb']
if freq_symptom in symptom_list:
    symptom_list.remove(freq_symptom)
symptom_list
```
