````python
import sys, json
sys.path.insert(0, '.')
from intake_engine.config import API_KEYS
print(repr(API_KEYS['anthropic'][:20]) + '...')  # just prints first 20 chars
````

````python
import anthropic
client = anthropic.Anthropic(api_key=API_KEYS['anthropic'])
response = client.messages.create(
    model='claude-haiku-4-5-20251001',
    max_tokens=4096,
    messages=[{'role': 'user', 'content': 'Return this JSON: {"test": true}'}]
)
print('Stop reason:', response.stop_reason)
print('Content:', repr(response.content))
````

````python
text = response.content[0].text.strip()
print('Raw text:', repr(text))
parsed = json.loads(text)
print('Parsed:', parsed)
````

````python
import re

def strip_fences(text: str) -> str:
    text = text.strip()
    text = re.sub(r'^```(?:json)?\s*', '', text)
    text = re.sub(r'\s*```$', '', text)
    return text.strip()

clean = strip_fences(response.content[0].text)
print('Clean text:', repr(clean))
parsed = json.loads(clean)
print('Parsed:', parsed)
````
